```python
#Input Section

#Carbon Nanotube Info
n=6              # chiral index n
m=6              # chiral index m
l=1              # length of the carbon nanotube in nm

#Funnel Info (Carbon Nanocone Info)
basediameter=15  #desired diameter of the carbon nanocone in Ångstrom
slevel=4         #smoothness/disclination level (1-5) referring to the disclination 
                 #angle of the carbon nanocone's graphene disk (1 = 60°, 2 = 120°, 
                 #3 = 180°, 4 = 240°, 5 = 300°)
```

```python
#Virtual Graphene Sheet Generation for Carbon Nanotube

if int((n+m+(l*10+2.46)//2.46)*2)>int(basediameter*100):
    maxringsonheight=int((n+m+(l*10+2.46)//2.46)*2)
else:
    maxringsonheight=int(basediameter*100)
    
if int((n+m+(l*10+2.84)//2.84)*2)>int(basediameter*100):
    maxringsonwidth=int((n+m+(l*10+2.84)//2.84)*2)
else:
    maxringsonwidth=int(basediameter*100)



maxonrow=[0,0,0]
maxonrow[0]=((maxringsonwidth+1)//2)*2
maxonrow[1]=maxringsonwidth+1
maxonrow[2]=(maxringsonwidth//2)*2
evenatoms=[]
oddatoms=[]
ycoord=round(0,2)
vPoints=[]
evenatoms.append(round((1.42*math.sin(math.radians(30))),2))
oddatoms.append(round((0),2))
for i1 in range(maxonrow[0]-1):
    if ((i1+1)%2==1):
        evenatoms.append(round((evenatoms[i1]+1.42),2))
    else:
        evenatoms.append(round((evenatoms[i1]+2*1.42*math.sin(math.radians(30))+1.42),2))
for i in range(len(evenatoms)):
    vPoints.append([evenatoms[i],ycoord])
for i2 in range(maxonrow[1]-1):
    if ((i2+1)%2==1):
        oddatoms.append(round((oddatoms[i2]+2*1.42*math.sin(math.radians(30))+1.42),2))
    else:
        oddatoms.append(round((oddatoms[i2]+1.42),2))
ycoord=round((ycoord+round((1.42*math.sin(math.radians(60))),2)),2)
for i in range(len(oddatoms)):
    vPoints.append([oddatoms[i],ycoord])
for i in range(int((maxringsonheight-1))):
    evenatoms=[]
    evenatoms.append(round((1.42*math.sin(math.radians(30))),2))
    oddatoms=[]
    oddatoms.append(round((0),2))
    ycoord=round((ycoord+round((1.42*math.sin(math.radians(60))),2)),2)
    for i2 in range(maxonrow[1]-1):
        if ((i2+1)%2==1):
            evenatoms.append(round((evenatoms[i2]+1.42),2))
            oddatoms.append(round((oddatoms[i2]+2*1.42*math.sin(math.radians(30))+1.42),2))
        else:
            evenatoms.append(round((evenatoms[i2]+2*1.42*math.sin(math.radians(30))+1.42),2))
            oddatoms.append(round((oddatoms[i2]+1.42),2))
    for i in range(len(evenatoms)):
        vPoints.append([evenatoms[i],ycoord])
    ycoord=round((ycoord+round((1.42*math.sin(math.radians(60))),2)),2)
    for i in range(len(oddatoms)):
        vPoints.append([oddatoms[i],ycoord])
evenatoms=[]
evenatoms.append(round((1.42*math.sin(math.radians(30))),2))
ycoord=round((ycoord+round((1.42*math.sin(math.radians(60))),2)),2)
for i2 in range(maxonrow[1]-1):
        if ((i2+1)%2==1):
            evenatoms.append(round((evenatoms[i2]+1.42),2))
        else:
            evenatoms.append(round((evenatoms[i2]+2*1.42*math.sin(math.radians(30))+1.42),2))
for i in range(len(evenatoms)):
    vPoints.append([evenatoms[i],ycoord])    
ycoord=round((ycoord+round((1.42*math.sin(math.radians(60))),2)),2)
oddatoms=[]
oddatoms.append(round((1.42+2*1.42*math.sin(math.radians(30))),2))
for i3 in range(maxonrow[2]-1):
    if ((i3+1)%2==1):
        oddatoms.append(round((oddatoms[i3]+1.42),2))
    else:
        oddatoms.append(round((oddatoms[i3]+2*1.42*math.sin(math.radians(30))+1.42),2))
for i in range(len(oddatoms)):
    vPoints.append([oddatoms[i],ycoord])
```

```python
#Preparation of the Carbon Nanotubes

#Vectors' components on the virtual graphene sheet
a1x=round(1.42+1.42*math.sin(math.radians(30)),2)
a1y=round(1.42*math.sin(math.radians(60)),2)
a2x=round(1.42+1.42*math.sin(math.radians(30)),2)
a2y=round(-1.42*math.sin(math.radians(60)),2)

# Zigzag Carbon Nanotube atom selection
if (n>m):
    originy=round(2*1.42*math.sin(math.radians(60)),2)
    Chy=round(originy+n*a1y+m*a2y,2)
    if maxringsonwidth%2==1:
        Chx=round(evenatoms[-3],2)
    else:
        Chx=round(evenatoms[-4],2)
    originx=round(Chx-n*a1x-m*a2x,2)
    Tx=originx-l*10*math.sin(math.radians(30))
    Ty=originy+l*10*math.sin(math.radians(60))
    ChTx=Chx-l*10*math.sin(math.radians(30))
    ChTy=Chy+l*10*math.sin(math.radians(60))
    #Line 1 origin-Ch
    slope1=(originy-Chy)/(originx-Chx)
    intercept1=(originx*Chy-Chx*originy)/(originx-Chx)
    #Line 2 origin-T
    slope2=(Ty-originy)/(Tx-originx)
    intercept2=(Tx*originy-originx*Ty)/(Tx-originx)
    #Line 3 T-ChT
    slope3=(Ty-ChTy)/(Tx-ChTx)
    intercept3=(Tx*ChTy-ChTx*Ty)/(Tx-ChTx)
    #Line 4 ChT-Ch
    slope4=(ChTy-Chy)/(ChTx-Chx)
    intercept4=(ChTx*Chy-Chx*ChTy)/(ChTx-Chx)
    
    PointsGtoCNT=[]
    for i in range(len(vPoints)):
        if (vPoints[i][0]>=(vPoints[i][1]-intercept2)/slope2) and (vPoints[i][0]-0.001<(vPoints[i][1]-intercept4)/slope4) and (vPoints[i][1]+0.001>=round((slope1*vPoints[i][0]+intercept1),2)) and (vPoints[i][1]<round((slope3*vPoints[i][0]+intercept3),2)):
            PointsGtoCNT.append(vPoints[i])

            
#Armchair Carbon Nanotube atom selection
elif (n==m):
    Chy=round(2*1.42*math.sin(math.radians(60)),2)
    originy=round(2*1.42*math.sin(math.radians(60)),2)            
    originx=round(1.42*math.sin(math.radians(30)),2)
    Chx=round(originx+n*a1x+m*a2x,2)
    Tx=(originy-Chy)+originx
    Ty=l*10+originy
    ChTx=(originy-Chy)+Chx
    ChTy=l*10+Chx
    #Line 1 origin-Ch
    slope1=0
    intercept1=0
    #Line 2 origin-T
    slope2='undefined'
    intercept2=0
    #Line 3 T-ChT
    slope3=0
    intercept3=0
    #Line 4 ChT-Ch
    slope4='undefined'
    intercept4=0
    
    PointsGtoCNT=[]
    for i in range(len(vPoints)):
        if (vPoints[i][0]>originx and vPoints[i][0]<=Chx and vPoints[i][1]>originy and vPoints[i][1]<=Ty):
            PointsGtoCNT.append(vPoints[i])

            
#Creation of the Carbon Nanotube            
PointsCNT=[]
aux=[]
for i in range(len(PointsGtoCNT)):
    CNTx=(math.sqrt((Chx-originx)**2+(Chy-originy)**2)/(2*math.pi)*math.cos(2*math.pi*((PointsGtoCNT[i][0]*(Chx-originx))+(PointsGtoCNT[i][1]*(Chy-originy)))/((Chx-originx)**2+(Chy-originy)**2)))
    CNTy=(math.sqrt((Chx-originx)**2+(Chy-originy)**2)/(2*math.pi)*math.sin(2*math.pi*((PointsGtoCNT[i][0]*(Chx-originx))+(PointsGtoCNT[i][1]*(Chy-originy)))/((Chx-originx)**2+(Chy-originy)**2)))
    CNTz=(PointsGtoCNT[i][0]*(Tx-originx)+PointsGtoCNT[i][1]*(Ty-originy))/math.sqrt((Tx-originx)**2+(Ty-originy)**2)
    auxx=round(CNTx,2)
    auxy=round(CNTy,2)
    auxz=round(CNTz,2)
    auxk=[auxx,auxy,auxz]
    if auxk not in aux:
        CNTxx=float(str(format(round(CNTx,3), '.3f')))
        CNTyy=float(str(format(round(CNTy,3), '.3f')))
        CNTzz=float(str(format(round(CNTz,3), '.3f')))
        aux.append(auxk)
        PointsCNT.append([CNTxx,CNTyy,CNTzz])

#Making the bottom attoms of the carbon nanotube start at the 0 value of the z-coord        
minzCNT=PointsCNT[0][2]
for i in range(len(PointsCNT)):
        if PointsCNT[i][2]<minzCNT:
            minzCNT=PointsCNT[i][2]
for i in range(len(PointsCNT)):
    if minzCNT<0:
        PointsCNT[i][2]=PointsCNT[i][2]+abs(minzCNT)
    elif minzCNT>=0:
        PointsCNT[i][2]=PointsCNT[i][2]-abs(minzCNT)
```

```python
#Calculation of CNT diameter, indexing the dangling and back bonds of CNT

CNTdiameter=math.sqrt(n**2+m**2+n*m)*2.46/math.pi
dbondsCNT=n+m         #dangling bonds
bondsperatomCNT=[]    #number of bonds each atom has
dbondsatomsCNT=[]     #identified atoms as dangling bonds at the bottom end of the carbon nanotube
orderedbondsrank=[]   #list of atoms and their respective bonds number
backbondsCNT=[]       #identified atoms as back bonds at the bottom end of the carbon nanotube
junctionatomsCNT=[]   #list of both the dangling bonds atoms and the back bonds atoms together, ordered

def takeSecond(elem): #ordering function for orderedbondsrank
    return(elem[1])

for i in range(len(PointsCNT)):
    bondsperatomCNT.append(0)
for i in range(len(PointsCNT)):
    for j in range(len(PointsCNT)):
        if i!=j:
            if (PointsCNT[i][0]-PointsCNT[j][0])**2+(PointsCNT[i][1]-PointsCNT[j][1])**2+(PointsCNT[i][2]-PointsCNT[j][2])**2<=1.422**2:
                bondsperatomCNT[i]=bondsperatomCNT[i]+1
                
for i in range(len(PointsCNT)):
    if bondsperatomCNT[i]<=2:
        orderedbondsrank.append([i,PointsCNT[i][2]])

orderedbondsrank.sort(key=takeSecond)
e=0
while (e<dbondsCNT):
    dbondsatomsCNT.append(orderedbondsrank[e][0])
    e=e+1
    
for i in range(dbondsCNT):
    for j in range(len(PointsCNT)):
        if dbondsatomsCNT[i]!=j:
            if (PointsCNT[dbondsatomsCNT[i]][0]-PointsCNT[j][0])**2+(PointsCNT[dbondsatomsCNT[i]][1]-PointsCNT[j][1])**2+(PointsCNT[dbondsatomsCNT[i]][2]-PointsCNT[j][2])**2<=1.422**2 and bondsperatomCNT[j]==3 and j not in backbondsCNT:
                backbondsCNT.append(j)

backbondsCNT.append(backbondsCNT[0])
del backbondsCNT[0]

if (n>m):
    par=0
    j=0
    k=0
    for i in range(len(dbondsatomsCNT)+len(backbondsCNT)):
        if par==0:
            junctionatomsCNT.append(dbondsatomsCNT[j])
            j=j+1
            par=1
        elif par==1:
            junctionatomsCNT.append(backbondsCNT[k])
            k=k+1
            par=0

if (n==m):
    par=0
    j=0
    k=0
    for i in range(len(dbondsatomsCNT)+len(backbondsCNT)):
        if par<2:
            junctionatomsCNT.append(dbondsatomsCNT[j])
            j=j+1
            par=par+1
        else:
            junctionatomsCNT.append(backbondsCNT[k])
            k=k+1
            par=par+1
            if par==4:
                par=0
```
